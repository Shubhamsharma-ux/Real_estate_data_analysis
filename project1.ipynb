```python
import pandas as pd 
from matplotlib import pyplot as plt
data=pd.read_csv("D:\\python2024\\estate4.csv")
months = "Month"
rural = "Profit_Percentage"
urban = "Average Commission per Flat"


# Ensure the 'Month' column is of type string
data[months] = data[months].astype(str)  # Convert 'Month' column to string
data['Month'] = pd.to_datetime(data['Month'])
yearly = data[(data['Month'].dt.year >=14)]


# Create a figure and axis
fig, ax = plt.subplots()

# Plot each line
ax.plot(data[months], data[rural], label='Profit_Percentage per month')
ax.plot(data[months], data[urban], label='Average Commission per Flat per month')

# Set titles and labels
ax.set_title('Profit and Commission Per Month')
ax.set_xlabel('Month')
ax.set_ylabel('profit and commission percentage')

# Add legend
ax.legend()

# Show grid
ax.grid(True)

# Display theplot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("D:\\python2024\\estate4.csv")
data['Month'] = pd.to_datetime(data['Month'])
months_data = data[(data['Month'].dt.year >= 14)]  
total_avg_sale = "Total Average Sale per Month"
avg_sale_1bhk = "Average Sale 1BHK per Month"
avg_sale_2bhk = "Average Sale 2BHK per Month"
avg_sale_3bhk = "Average Sale 3BHK per Month"

# Create a figure and axis
fig, ax = plt.subplots()

# Plot each line
ax.plot(months_data['Month'], months_data[total_avg_sale], label='Total Average Sale')
ax.plot(months_data['Month'], months_data[avg_sale_1bhk], label='Average Sale 1BHK')
ax.plot(months_data['Month'], months_data[avg_sale_2bhk], label='Average Sale 2BHK')
ax.plot(months_data['Month'], months_data[avg_sale_3bhk], label='Average Sale 3BHK')

# Set titles and labels
ax.set_title('NSPL REAL ESTATE Sales Data From Year 14 Onwards')
ax.set_xlabel('Month')
ax.set_ylabel('Sales Amount')

# Add legend
ax.legend()
# Show grid
ax.grid(True)

# Display theplot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np



data=pd.read_csv("D:\\python2024\\estate4.csv")
months = "Month"
complition_of_1bhk = "Completion of 1BHK Flats per Month"
complition_of_2bhk = "Completion of 2BHK Flats per Month"
complition_of_3bhk = "Completion of 3BHK Flats per Month"
complition_of_4bhk= "Completion of 4BHK Flats per Month"

# Ensure the 'Month' column is of type string
data[months] = data[months].astype(str)  # Convert 'Month' column to string

# Create a figure and axis
fig, ax = plt.subplots()

# Plot each line
ax.plot(data[months], data[complition_of_1bhk], label='complition of 1bhk')
ax.plot(data[months], data[complition_of_2bhk], label='complition of 2bhk')
ax.plot(data[months], data[complition_of_3bhk], label='complition of 3bhk')
ax.plot(data[months], data[complition_of_4bhk], label='complition of 4bhk')

# Set titles and labels
ax.set_title('NSPL REAL ESTATE Completion Period For Different Types Of Flats')
ax.set_xlabel('Month')
ax.set_ylabel('completion of flats')

# Add legend
ax.legend()

# Show grid
ax.grid(True)

plt.xlim(0,6)
# Display theplot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("D:\\python2024\\real_estate_profit_percentage_2014_2023_percentage3.csv")

customer_counts_1bhk = data[data["flat preference"] == "1BHK"].groupby('number of customers')['flat preference'].count()
customer_counts_2bhk = data[data["flat preference"] == "2BHK"].groupby('number of customers')['flat preference'].count()
customer_counts_3bhk = data[data["flat preference"] == "3BHK"].groupby('number of customers')['flat preference'].count()


plt.figure(figsize=(10, 6)) 

plt.plot(customer_counts_1bhk.index, customer_counts_1bhk.values, label='1BHK')
plt.plot(customer_counts_2bhk.index, customer_counts_2bhk.values, label='2BHK')
plt.plot(customer_counts_3bhk.index, customer_counts_3bhk.values, label='3BHK')

plt.xlabel('Number of Customers')
plt.ylabel('Count same values')
plt.title('NSPL REAL ESTATE On Customer Counts for Different Flat Preferences')

plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("D:\\python2024\\real_estate_profit_percentage_2014_2023_percentage3.csv")

customer_counts_lt50k = data[data["Salary range"] == "<50k"].groupby('number of customers')['Salary range'].count()
customer_counts_50k100k = data[data["Salary range"] == "50k-100k"].groupby('number of customers')['Salary range'].count()
customer_counts_gt100k = data[data["Salary range"] == ">100k"].groupby('number of customers')['Salary range'].count()

plt.figure(figsize=(10, 6)) 

plt.plot(customer_counts_lt50k.index, customer_counts_lt50k.values, color="black",label='<50k')
plt.plot(customer_counts_50k100k.index, customer_counts_50k100k.values, label='50k-100k')
plt.plot(customer_counts_gt100k.index, customer_counts_gt100k.values, label='>100k')

plt.xlabel('Number of Customers')
plt.ylabel('Count')
plt.title('NSPL REAL ESTATE ON Customer Counts for Different Salary Ranges')

plt.legend()

plt.show()
```
